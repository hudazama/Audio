```python
import pandas as pd
import numpy as np
import librosa
import os
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import glob
import librosa.display
import IPython.display as ipd
```

```python
glob.glob('./Vocal_Audio_Dir/*.wav')
```

```python
wavfile = glob.glob('./Vocal_Audio_Dir/*.wav')[-2]
ipd.Audio(wavfile)
```

```python
data, sr = librosa.load(wavfile,44100)
print('the file shape is', data.shape)
print('the sampling rate is',sr)
plt.figure(figsize=(14, 5))
librosa.display.waveplot(data)
plt.title('Monophonic')
```

```python
data, sr = librosa.load(wavfile,44100,mono=False)
print('the file shape is', data.shape)
print('the sampling rate is',sr)
plt.figure(figsize=(14, 5))
librosa.display.waveplot(data)
plt.title('Stereo')
```
